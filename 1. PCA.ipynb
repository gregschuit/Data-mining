# Principal Component Analisis

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
mean = [2, 3, 5]
```

```python
precov = [[2, 3, 0],
       [1, 2, 1], 
       [1, 1, 2]]
```

```python
cov = np.dot(precov, np.transpose(precov))
```

```python
x, y, z = np.random.multivariate_normal(mean, cov, 300).T
x = [(d - min(x)) / (max(x) - min(x)) for d in x]
y = [(d - min(y)) / (max(y) - min(y)) for d in y]
z = [(d - min(z)) / (max(z) - min(z)) for d in z]

plt.plot(x, y, 'x')
#plt.axis('equal')
plt.show()
```

```python
fig = plt.figure()
puntos = fig.add_subplot(111, projection='3d')
puntos.scatter(x, y, z)
```

```python
matriz_datos = np.vstack([x,y,z])
cov_empirica = np.cov(matriz_datos)
```

```python
cov_empirica
```

```python
values, vectors = np.linalg.eig(cov_empirica)
values
```

```python
puntos.quiver(vectors[0])
plt.show()
```
